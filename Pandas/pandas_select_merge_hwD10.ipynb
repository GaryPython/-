```python
#讀取STOCK_DAY_0050_202009.csv、STOCK_DAY_0050_202010.csv
data_09 = pd.read_csv("STOCK_DAY_0050_202009.csv")
data_10 = pd.read_csv("STOCK_DAY_0050_202010.csv")
print("STOCK_DAY_0050_202009.csv\n\n",  data_09, "\n\nSTOCK_DAY_0050_202010.csv\n\n",  data_10)
```

```python
total = len(data_09) + len(data_10)
total
np.arange(total)
```

```python
#串連
```

```python
pd.merge(data_09, data_10,how='outer')
```

```python
#串聯
datas = pd.concat([data_09,data_10],axis=0)
datas.set_index(np.arange(total))
```

```python
#找出open小於close的資料
```

```python
datas.loc[(datas.open<datas.close)]
```
