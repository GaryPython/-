# 散點圖: Scatter Plots
    
    顏色由（X，Y）的角度給出。

    注意標記的大小，顏色和透明度。

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

'''
#作業

#方法不只有一種

'''

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.scatter(X,Y,alpha=0.3 ,c='r')

plt.show()
```
